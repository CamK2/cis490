```python
img = cv2.imread('Lenna_(test_image).png')
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(img)
plt.imshow(img2)
plt.show()
```

```python
img_me = cv2.imread('5ef22a433c1f2.image.jpg?resize=500,750')
img_me2 = cv2.cvtColor(img_me, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.imshow(img_me)
plt.imshow(img_me2)
plt.show()
```

```python
# plot just one channel
plt.imshow(img[:,:,2],cmap='gray')
plt.show()

# plot just one channel
plt.imshow(img[:,:,2],cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
blur_img = cv2.GaussianBlur(img_me2, (15,15), 7)
plt.imshow(blur_img)
plt.show()
```

```python
edges = cv2.Canny(img_me2,50,300)
plt.imshow(edges)
plt.show()
```

```python
logo_img = cv2.imread("OpenCV_logo_black_.png")
plt.imshow(logo_img)
logo_copy = logo_img.copy

circles = cv2.HoughCircles(logo_img,cv2.HOUGH_GRADIENT,0.1, 50, param1=50, param2=38)
for x, y, r in circles[0,:]:
  cv2.circle(logo_img)


plt.show()
```
